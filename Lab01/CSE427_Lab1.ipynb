```python
import numpy as np
```

```python
#1
a = np.random.randint(1, 101, 20)
b = np.random.randint(1, 101, 20)
print(f"Array a: {a}")
print(f"Array b: {b}")
```

```python
#2
a_matrix = a.reshape(5, 4)
b_matrix = b.reshape(5, 4)
print(f"Matrix a:\n{a_matrix}")
print(f"Matrix b:\n{b_matrix}")
```

```python
#3
scalar_multiplication = a_matrix * b_matrix
print(f"Scalar Multiplication:\n{scalar_multiplication}")
```

```python
#4
b_modified = b_matrix.reshape(4, 5)
c = a_matrix @ b_modified
print(f"Matrix b:\n{b_modified}")
print(f"Matrix c:\n{c}")
```

```python
#5
print(f"Columns 2 and 3 of matrix C:\n{c[:, 1:3]}")
print(f"matrix c using slicing:\n{c[:, :]}")
```

```python
#6
largest_element = np.max(c)
index_of_largest = np.unravel_index(np.argmax(c), c.shape)
print(f"Largest element in matrix c: {largest_element}")
print(f"largest element idx: {index_of_largest}")
```

```python
#7
c_oned = c.flatten()
print(f"Matrix c as a 1D array:\n{c_oned}")
print(f"Original matrix c remains unchanged:\n{c}")
```

```python
import pandas as pd
```

```python
#1
csv_path = '/content/subject_scores.csv'
df = pd.read_csv(csv_path)
```

```python
#2
df.head()
```

```python
#3
df.info()
df.describe()
```

```python
#4
df.isnull().sum()
```

```python
#5
numeric_cols = df.select_dtypes(include=np.number).columns
for col in numeric_cols:
    mean_val = df[col].mean()
    df[col] = df[col].fillna(mean_val)
df.isnull().sum()
```

```python
#6
math_score_col_name = [col for col in df.columns if 'Math' in col or 'math' in col][0]
print(f"{math_score_col_name} of all students:")
print(df[math_score_col_name])
```

```python
import matplotlib.pyplot as plt
```

```python
#1
temperature_dhaka = [25,34,21,45,28,6,43,18,7,2]
humidity_dhaka = [28, 25,29,20, 26, 50, 19, 29, 52, 55]
plt.figure(figsize=(8, 6))
plt.scatter(temperature_dhaka, humidity_dhaka, marker='*', color='red', label='Dhaka Data')
plt.title('Dhaka: Temperature vs Humidity Scatter Plot')
plt.xlabel('Temperature (Â°C)')
plt.ylabel('Humidity (%)')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

```python
#2
study_hours = [2,3,4,4, 5, 6, 7, 7, 8, 9, 9, 10, 11, 11, 12]
marks = [6, 10, 15, 20, 34, 44, 55, 60, 55, 67, 70, 80, 90, 99, 100]
plt.figure(figsize=(12, 8))
plt.plot(study_hours, marks, marker='o', linestyle='-', color='blue')
plt.title('Study Hours vs Marks', fontsize=16)
plt.xlabel('Study Hours', fontsize=14)
plt.ylabel('Marks', fontsize=14)
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

```python
#3
subjects = ['Maths', 'English', 'Science', 'Physics', 'Computer']
marks = [89, 90, 45, 78, 99]
plt.figure(figsize=(10, 6))
plt.barh(subjects, marks, color='skyblue')
plt.title('Student Scores in Subjects (Horizontal Bar)')
plt.xlabel('Marks Scored')
plt.ylabel('Subjects')
plt.xlim(0, 100)
plt.grid(axis='x', linestyle='--', alpha=0.6)
plt.show()
```

```python
#4
plt.figure(figsize=(10, 6))
colors = ['red', 'green', 'blue', 'orange', 'purple']
plt.bar(subjects, marks, color=colors)
plt.title('Student Scores in Subjects (Vertical Bar with Different Colors)')
plt.xlabel('Subjects')
plt.ylabel('Marks Scored')
plt.ylim(0, 100)
plt.grid(axis='y', linestyle=':', alpha=0.7)
plt.show()
```

```python
#5
plt.figure(figsize=(12, 8))
plt.subplot(1, 2, 1)
plt.barh(subjects, marks, color='skyblue')
plt.title('Horizontal Bar Graph: Subject Scores', fontsize=14)
plt.xlabel('Marks')
plt.ylabel('Subjects')
plt.grid(True, axis='x', linestyle='--', alpha=0.6)
plt.subplot(1, 2, 2)
colors = ['red', 'green', 'blue', 'orange', 'purple']
plt.bar(subjects, marks, color=colors)
plt.title('Vertical Bar Graph: Subject Scores', fontsize=14)
plt.xlabel('Subjects')
plt.ylabel('Marks')
plt.grid(True, axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```
